```python
df.iloc[:,10]  # 대표국적 추리기
```

```python
n1 = df.iloc[:,10].value_counts() # 대표국적 카운트
n1
```

```python
n1d = n1.to_dict()
plt.figure(figsize=(8, 8))
plt.pie(n1d.values(), labels=n1d.keys(), autopct='%1.1f%%')
plt.axis('equal')  # 동그란 원 형태로 표시
plt.title('대표국적 분포')
plt.show()
```

```python
n2=n1.iloc[0:5]  # 대표국적 상위 5개, 기타 국가를 하나로 묶는방법?
n2
```

```python
n2d = n2.to_dict()
n2d
```

```python
plt.figure(figsize=(8, 8))
plt.pie(n2d.values(), labels=n2d.keys(), autopct='%1.1f%%')
plt.axis('equal')  # 동그란 원 형태로 표시
plt.title('대표국적 분포')
plt.show()
```
